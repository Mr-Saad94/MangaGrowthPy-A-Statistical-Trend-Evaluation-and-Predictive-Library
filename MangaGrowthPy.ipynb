```python
%%writefile mangagrowthpy/metrics.py

def calculate_mean(data):
    return sum(data) / len(data)

def calculate_standard_deviation(data):
    mean = calculate_mean(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    return variance ** 0.5

def calculate_standard_error(data):
    sd = calculate_standard_deviation(data)
    return sd / (len(data) ** 0.5)

def confidence_interval_95(data):
    mean = calculate_mean(data)
    se = calculate_standard_error(data)
    margin = 1.96 * se
    lower = mean - margin
    upper = mean + margin
    return lower, upper
```

```python
%%writefile mangagrowthpy/prediction.py

#Growth&Prediction Functions

def growth_rate(data):
    if len(data) < 2:
        return 0
    return ((data[-1] - data[-2]) / data[-2]) * 100


def predict_next_month(last_value, growth_percent):
    return last_value * (1 + growth_percent / 100)


def chapter_consistency(total_chapters, months_active):
    if months_active == 0:
        return 0
    return total_chapters / months_active


def success_category(score):
    if score >= 80:
        return "High Potential 🚀"
    elif score >= 50:
        return "Moderate Potential 📈"
    else:
        return "Low Potential ⚠️"
```

```python
from mangagrowthpy import *

print("----- Welcome to MangaGrowthPy -----")

name = input("Enter Manga Name: ")
rating = float(input("Enter Rating (0-10): "))
chapters = int(input("Enter Total Chapters: "))
months_active = int(input("Enter Months Since Release: "))

views_input = input("Enter Monthly Views separated by commas: ")
views = [int(v.strip()) for v in views_input.split(",")]

#Calculations

mean_views = calculate_mean(views)
sd = calculate_standard_deviation(views)
ci_lower, ci_upper = confidence_interval_95(views)

growth = growth_rate(views)
next_month_prediction = predict_next_month(views[-1], growth)

consistency = chapter_consistency(chapters, months_active)

# Simple Popularity Score Formula
popularity_score = (rating * 10) + (growth * 0.5) + (consistency * 5)

category = success_category(popularity_score)

print("\n----- MangaGrowthPy Analytical Report -----")
print("Manga:", name)
print("Average Monthly Views:", round(mean_views, 2))
print("Standard Deviation:", round(sd, 2))
print("95% Confidence Interval:", (round(ci_lower,2), round(ci_upper,2)))
print("Growth Rate:", round(growth, 2), "%")
print("Predicted Next Month Views:", round(next_month_prediction, 2))
print("Chapter Consistency:", round(consistency, 2))
print("Popularity Index:", round(popularity_score, 2))
print("Category:", category)
```
